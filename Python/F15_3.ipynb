```
import numpy as np
import matplotlib.pyplot as plt
R= np.arange(0, 0.2, 0.002)

def R_Plus_1(R):
    return 1+R

def R_Plus_1_Exp(R, exp):
    return R_Plus_1(R)**exp

def Price(nYear, R, payPerYear):
    PriceAccumulate = 0
    for year in range(1,nYear+1):
        PriceAccumulate = PriceAccumulate + payPerYear/(R_Plus_1_Exp(R,year))
    PriceAccumulate = PriceAccumulate + 1/R_Plus_1_Exp(R,nYear) # 1 is 1 million for the scrap value
    return PriceAccumulate

nNumberOfYear = 20
profitPerYear = 0.96 #million dolar // havn't less the investCapital
investCapital = 10 #million dolar
Revenue = Price(nNumberOfYear, R, profitPerYear)  # idn "Revenue" is it suitable name.
netPresenValue = Revenue-investCapital

# calculate 0 NPV at which R*
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

zeroNPV = 0
nearestZeroNPV = find_nearest(netPresenValue, zeroNPV)
idxZeroNPV = np.where(netPresenValue==nearestZeroNPV)
R_AtZeroNPV = R[idxZeroNPV]
plt.plot(R, netPresenValue)
plt.vlines(R_AtZeroNPV, -6, 11)
print(R_AtZeroNPV)
plt.show()
```
